### 4
Прочитайте у файлі `DataDictionary-ua.txt`, що означають стовпчики матриці. Якому типу належить кожен стовпчик (дійсний, цілий, категоріальний)?

```python
import re
with open('DataDictionary-ua.txt', 'r') as f:
    contents = f.read()
columns = re.findall(r'(\w+)\s+(.+)', contents)

for column in columns:
    name, description = column
    if 'integer' in description:
        print(f"{name}: Цілий (integer)")
    elif 'real' in description:
        print(f"{name}: Дійсний (real)")
    elif '%' in description:
        print(f"{name}: Дійсний (відсотки)")
    elif 'Y/N' in description:
        print(f"{name}: Категоріальний (Y/N)")
```

### 8

Використовуйте метод `groupby`, оцініть ймовірність неповернення кредиту (`SeriousDlqin2yrs=1`) для різних кількості утриманців (`NumberOfDependents`).  

Проробіть аналогічну процедуру для різних значень стовпчика `NumberRealEstateLoansOrLines`

*Підказка:*
`data['стовпчик1'].groupby(data['стовпчик2']).mean()`  *-- розрахунок середніх значень стовпчика1 по групам зі стовпчика2*

```python
dep_prob = data.groupby('NumberOfDependents')['SeriousDlqin2yrs'].mean()

# Виведення результатів
print(dep_prob)

# Оцінка ймовірності неповернення кредиту для різних значень стовпчика NumberRealEstateLoansOrLines
loan_prob = data.groupby('NumberRealEstateLoansOrLines')['SeriousDlqin2yrs'].mean()

# Виведення результатів
print(loan_prob)
```

### 9a

Побудуйте графік розсіювання на вісях `age` та `Debt`. Cинім відміттє клієнтів без серйозних заборгованостей (`SeriousDlqin2yrs = 0`) та червоним - боржників (`SeriousDlqin2yrs = 1`).

```python
no_dq = data[data['SeriousDlqin2yrs'] == 0]
dq = data[data['SeriousDlqin2yrs'] == 1]

# побудова графіку
plt.scatter(no_dq['age'], no_dq['Debt'], color='blue', alpha=0.3, label='No Delinquency')
plt.scatter(dq['age'], dq['Debt'], color='red', alpha=0.3, label='Delinquency')
plt.ylim(0, 2)
plt.xlabel('Age')
plt.ylabel('Debt')
plt.legend()
plt.show()
```

### 9b
Побудуйте на однрму графіку дві **нормовані** щільності розподілення: червону – для місяіного доходу клієнтів з заборгованостями, синю - для місяіного доходу клієнтів без заборгованостей. По вісі абсцис відобразіть значення до 25000.

```python
df_default = data[data['SeriousDlqin2yrs'] == 1]
df_no_default = data[data['SeriousDlqin2yrs'] == 0]

# Побудова графіку
plt.hist(df_default['MonthlyIncome'], bins=100, density=True, alpha=0.5, color='red', label='З заборгованістю')
plt.hist(df_no_default['MonthlyIncome'], bins=100, density=True, alpha=0.5, color='blue', label='Без заборгованості')
plt.xlim(0, 25000)
plt.xlabel('Місячний дохід')
plt.ylabel('Щільність розподілу')
plt.title('Розподіл місячного доходу за наявності заборгованості')
plt.legend(loc='best')
plt.show()
```
